```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Loading the Dataset from the URL
url = 'https://raw.githubusercontent.com/salemprakash/EDA/main/Data/countymurders.csv'
df = pd.read_csv(url)
print("First 5 rows of the dataset:")
print(df.head())
```

```python
# Summary Statistics for Numerical and Categorical Data
print("\nSummary statistics for numerical data:")
print(df.describe())  # Summary for numerical features
```

```python
# Checking for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("\nCategorical Columns:")
print(categorical_cols)
```

```python

# Summary for categorical columns
if len(categorical_cols) > 0:
    print("\nSummary for categorical columns:")
    print(df[categorical_cols].describe())
```

```python
#Data Cleaning (Handling Missing Values)
print("\nMissing values in the dataset:")
print(df.isnull().sum())
```

```python
# Dropping rows with missing data
df_cleaned = df.dropna()
print("\nData after dropping missing values:")
print(df_cleaned.head())
```

```python
# Filling missing data
df_filled = df.fillna(method='ffill')
print("\nData after forward fill:")
print(df_filled.head())
```

```python
plt.figure(figsize=(8, 6))
df['murdrate'].dropna().hist(bins=20)
plt.title('Histogram of Murder Rates')
plt.xlabel('Murder Rate')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot for a categorical variable (if available)
if len(categorical_cols) > 0:
    plt.figure(figsize=(8, 6))
    df[categorical_cols[0]].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {categorical_cols[0]}')
    plt.xlabel(categorical_cols[0])
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 5. Measurement Scales (Nominal, Ordinal, Interval, Ratio)
# For Nominal (e.g., city names):
if 'city' in df.columns:
    print("\nValue counts for 'city' column (Nominal scale example):")
    print(df['city'].value_counts())
```

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import KBinsDiscretizer
```

```python
#Data Deduplication
print("\nShape before deduplication:", df.shape)
df_deduped = df.drop_duplicates()
print("Shape after deduplication:", df_deduped.shape)
```

```python
#Replacing Values
# Replace specific values in a column
df_replaced = df.copy()
df_replaced['murders'] = df_replaced['murders'].replace(0, np.nan)

print("\n'Murders' column after replacing 0 with NaN:")
print(df_replaced[['murders']].head())
```

```python
# Replace NaN values with a specific value
df_replaced.fillna(value={'arrestrate': df_replaced['arrestrate'].mean()}, inplace=True)
print("\nSales column after replacing NaN with mean value:")
print(df_replaced[['arrestrate']].head())
```

```python
# 4. Discretization and Binning
# Bin 'median' column into categoriesdf_binned = df.copy()
df_binned['murders_filled'] = df_binned['murders'].fillna(df_binned['murders'].mean())
df_binned['murders_binned'] = pd.cut(df_binned['murders_filled'], bins=5, labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'])
print("\nBinned 'murders' column:")
print(df_binned[['murders', 'murders_binned']].head())
```

```python
# 5. Handling Missing Data - Traditional Method (Maximum Likelihood Estimation)
# Introduce some missing values for demonstration
df_with_na = df.copy()
df_with_na.loc[10:15, 'arrestrate'] = np.nan
# Handle missing data using SimpleImputer (Maximum Likelihood Estimation approximated via mean substitution here)
imputer = SimpleImputer(strategy='mean')
df_with_na['arrestrate'] = imputer.fit_transform(df_with_na[['arrestrate']])
print(df_with_na[['arrestrate']].head(20))
```

```python
from sklearn.preprocessing import KBinsDiscretizer
import numpy as np

# Copy dataset
df_binned = df.copy()

# Fill NaN values in 'murders' with the mean before discretization
df_binned['murders_filled'] = df_binned['murders'].fillna(df_binned['murders'].mean())

# Apply KBinsDiscretizer
kbins = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')  # 'uniform' bins
df_binned['murders_binned_kbins'] = kbins.fit_transform(df_binned[['murders_filled']])

# Display results
print("\nDiscretized 'murders' column using KBinsDiscretizer:")
print(df_binned[['murders', 'murders_binned_kbins']].head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
df_binned["density"].value_counts().sort_index().plot(kind='bar')
plt.title('Distribution of arrests ')
plt.xlabel("density")
plt.ylabel('Frequency')
plt.show()
```

```python
print("Univariate Analysis (Summary Statistics):")
print(df.describe())

# Visualizing Univariate Data (Histogram for 'murders')
plt.figure(figsize=(10, 6))
df['murders'].hist(bins=20)
plt.title('Histogram of Murders')
plt.xlabel('Number of Murders')
plt.ylabel('Frequency')
plt.show()
```

```python
print("\nBivariate Analysis (Correlation Matrix):")
# Select only numerical columns for correlation calculation
correlation_matrix = df.select_dtypes(include=np.number).corr()
print(correlation_matrix)

# Heatmap of Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()

# Scatter Plot for Bivariate Analysis (Example: Murders vs Murdrate)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='murders', y='murdrate', data=df)
plt.title('Scatter Plot of Murders vs Murder Rate')
plt.xlabel('Murders')
plt.ylabel('Murder Rate')
plt.show()
```

```python
# Convert 'year' column to datetime (not necessary here, but you can create a 'date' column for a more detailed analysis)
df['date'] = pd.to_datetime(df['year'], format='%Y')

# Time-based indexing (set 'date' as the index)
df.set_index('date', inplace=True)
print("\nDataset with time-based index:")
print(df.head())

# Visualizing Time Series Data (Murders over Time)
plt.figure(figsize=(10, 6))
df['murders'].plot()
plt.title('Murders Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Murders')
plt.show()

# Grouping Time Series Data (Yearly Total Murders)
yearly_murders = df.groupby(df.index.year)['murders'].sum()

plt.figure(figsize=(10, 6))
yearly_murders.plot(kind='bar')
plt.title('Yearly Murders')
plt.xlabel('Year')
plt.ylabel('Total Murders')
plt.show()

# Resampling Time Series Data (Monthly Total Murders)
monthly_murders = df['murders'].resample('M').sum()

plt.figure(figsize=(10, 6))
monthly_murders.plot()
plt.title('Monthly Murders Over Time')
plt.xlabel('Date')
plt.ylabel('Monthly Murders')
plt.show()

# Resampling for Quarterly Data (Total Murders)
quarterly_murders = df['murders'].resample('Q').sum()

plt.figure(figsize=(10, 6))
quarterly_murders.plot(kind='bar')
plt.title('Quarterly Murders Over Time')
plt.xlabel('Date')
plt.ylabel('Quarterly Murders')
plt.show()
```
